```python
import random
import matplotlib.pyplot as plt
import numpy as np
import time
from datetime import datetime
import pandas as pd
plt.style.use('fivethirtyeight')
pd.options.mode.chained_assignment = None  # default='warn'
import seaborn as sns
import math as mh
```

```python
A = []
for i in range(1, 2001):
  for j in range(1, 2001):
    A.append([i, j])
def f(x, y):
  return(-(x-3)**2 - (y+2)**2)
A = np.array(A)
```

```python
answer = np.array([np.nan]*4000000).reshape((2000, 2000))
```

```python
S = np.array([1, 2, 3, 4]).reshape((2, 2))
S[0]
```

```python

for t in range(1, 5):
  if t > 1:
    y = 2000 - np.nanargmax(answer.reshape(1, -1)) // 2000
    x = np.nanargmax(answer.reshape(1, -1)) % 2000
    A = A[A[:,0] >= x - (2000 // (2**t))]
    A = A[A[:,0] <= x + (2000 // (2**t))]
    A = A[A[:,1] >= y - (2000 // (2**t))]
    A = A[A[:,1] <= y + (2000 // (2**t))]
  L = random.sample(list(A), len(A) // (100 // 2**(t-1)))
  Step_X = np.array(L)[:, 0]
  Step_Y = np.array(L)[:, 1]
  for i in range(len(L)):
      answer[2000 - Step_Y[i]-1][Step_X[i]-1] = f((Step_X[i]-1)*0.01-10, (Step_Y[i]-1)*0.01-10)
  plt.figure(figsize=(10, 10))
  ax = sns.heatmap(answer, cmap = 'Greens')
  ax.set_title('Шаг цикла: ' + str(t) + ', процент протестированных комбинаций: ' + str(round(100 - sum(sum(np.isnan(answer))) / 40000, 4)))
  ax.set_xticks(np.arange(0, 2100, 100))
  ax.set_xticklabels(range(-10, 11), rotation = 0)
  ax.set_yticks(np.arange(0, 2100, 100))
  ax.set_yticklabels(range(10, -11, -1), rotation = 0)
  ax.plot(1300, 1200, 'ro')
  ax.grid()
  plt.show()
```
